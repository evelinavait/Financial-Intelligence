2.   Teigiamų elementų anuliavimas

```python
# Teigiamų elementų anuliavimas, parašytas su PYTHON, naudojant ciklą:
a = np.random.randint(low=-10, high=10, size=10)
print(a)

for i, val in enumerate(a):
  if val > 0:
    a[i] = 0
print(a)


# Teigiamų elementų anuliavimas, parašytas su PYTHON be ciklų:
a[a > 0] = 0
print(a)

# Teigiamų elementų anuliavimas, parašytas su PYTHON be ciklų (nekeičiant pradinio array):
a = np.random.randint(low=-10, high=10, size=10)
print(a)
out = np.where(a > 0, 0, a)
print(out)
print(a)

# numpy.where(condition, [x, y, ]/)
# condition: sąlyga
# x: reikšmė, kuri bus priskirta tiems elementams, kurie tenkina sąlygą
# y: reikšmė, kuri bus priskirta tiems elementams, kurie netenkina sąlygos
```

3.   Išmetimas > 6



```python
# Išmetimas > 6, parašytas su PYTHON, naudojant ciklą:
a = np.random.randint(low=1, high=10, size=10)
b = []
for i, val in enumerate(a):
  if val <= 6: # tikrinama, ar val yra mažesnis arba lygus 6
    b = np.append(b, val) # pridedama prie sąrašo b
b = b.astype(int) # verčiama į sveikuosius skaičius
print(a)
print(b)


# Išmetimas > 6, parašytas su PYTHON be ciklų (1 var):
b = a[a <= 6]
# b = b.astype(int)
print(type(b[0])) # Pirmo elemento tipas
print(b)

# Išmetimas > 6, parašytas su PYTHON be ciklų (2 var):
b = a[~(a > 6)]
print(b)
# a > 6 grąžina True reikšmes, kur a masyvo elementai yra > 6
# naudojant ~ reikšmės konvertuojamos į False -> atrenkami elementai, kurie <= 6

# Išmetimas > 6, parašytas su PYTHON be ciklų (3 var):
index = np.where(a <= 6) # indeksai, atitinkantys sąlygą
print(index) # spausdinamas indeksų masyvas
b = a[index] # elementai pagal nurodytus indeksus
print(b)
```

4.   Dviejų vienodų šalia esančių skaičių radimas



```python
# Kodas MATLAB formatu

# disp('-----------')
# a = randi(5,1,10);
# for i=2:length(a)
#     if a(i) == a(i-1)
#         disp(i);
#     end
# end

# Dviejų vienodų šalia esančių skaičių radimas, naudojant ciklą:
print("-----------")
a = np.random.randint(1, 6, 10) # nuo 1 iki 5 (neįskaitant 6)
print(a)

for i in range(1, len(a)):
    if a[i] == a[i - 1]:
      print(f"Indeksas: {i-1}, Reikšmė: {a[i]}")

# Dviejų vienodų šalia esančių skaičių radimas be ciklo:
# Palyginimas su sąrašo narysiais, kurie yra šalia
b = np.where(a[1:] == a[:-1])
print(b[0]) # grąžinamas pirmasis indeksas
# a[1:] nuo indekso 1 iki galo
# a[:-1] nuo pradžios iki paskutinio elemento


# Kitas būdas:
print("-----------")
a = np.random.randint(1, 6, 10) # nuo 1 iki 5 (neįskaitant 6)
print(a)

indices = np.where(np.diff(a) == 0)[0]
for i, j in zip(indices, indices + 1):
    print(f"Indeksai: {i}, {j}, Reikšmės: {a[i]}, {a[j]}")
```

6.   Elementų perstūmimas vektoriuje pakartojant paskutinį



```python
# Kodas MATLAB formatu

# disp('-----------')
# a = randi(10,1,10);
# disp(a);
# for i=2:length(a)
#     a(i-1) = a(i);
# end
# disp(a);

# Elementų perstūmimas vektoriuje pakartojant paskutinį, naudojant ciklą:
print("-----------")
a = np.random.randint(low=1, high=10, size=10)
print(a)
for i in range(1, len(a)):
    a[i - 1] = a[i] # reikšmės masyve perkeliamos vienu indeksu į kairę
print(a)

# Elementų perstūmimas vektoriuje pakartojant paskutinį be ciklų:
print("-----------")
a = np.random.randint(low=1, high=10, size=10)
print(a)
a = np.roll(a, -1) # elementai perstumiami viena pozicija į kairę (paskutinis elementas tampa pirmuoju)
a[-1] = a[-2]  # paskutinio elemento reikšmė = priešpaskutinio elemento reikšmei
print(a)

print("-----------")
a = np.random.randint(low=1, high=10, size=10)
print(a)
a[:-1] = a[1:]  # reikšmės iki paskutinės (išskyrus paskutinę) bus priskirtos reikšmėms nuo antros iki paskutinės
print(a)


print("-----------")
a = np.random.randint(low=1, high=10, size=10)
print(a)

size = len(a)
last = a[size - 1] # paskutinis elementas vektoriuje

a = np.delete(a,0) # pašalina 0-inio indekso elementą (pirmąjį)

new_a = np.append(a, last)
print(new_a)
```

2.   Minutiniai duomenys

```python
# https://medium.com/@kasperjuunge/yfinance-10-ways-to-get-stock-data-with-python-6677f49e8282
import pandas as pd
import yfinance as yf

def minute_data(symbol, start_date, end_date, output_file, date_format='%Y-%m-%d %H:%M:%S%z'):

    minute_data = yf.download(symbol, start=start_date, end=end_date, interval="1m")

    # Išsaugoma į CSV failą
    minute_data.to_csv(output_file)

    # Failas nuskaitomas į DataFrame
    data = pd.read_csv(output_file)

    # 'Datetime' stulpelis paverčiamas datetime formatu
    data['Datetime'] = pd.to_datetime(data["Datetime"], format=date_format)

    # 'Datetime' stulpelis nustatomas kaip indeksas
    data.set_index('Datetime', inplace=True)

    # Indeksas paverčiamas datetime formatu
    data.index = pd.to_datetime(data.index)

    # Formatuojamas indeksas kaip 'string'
    data.index = data.index.strftime('%Y-%m-%d %H:%M:%S')

    # Indeksas paverčiamas datetime formatu po pavertimo į 'string'
    minute_data.index = pd.to_datetime(minute_data.index)

    return data

minute_data_AAPL = minute_data("AAPL", "2024-02-13", "2024-02-14", "minute_data_AAPL.csv")
minute_data_AAPL
```

# 1.4. Atsitiktinių duomenų generavimas ir atvaizdavimas - generuoti ir atvaizduoti duomenis, kurie vizualiai būtų panašūs į rinkos kainų kitimo duomenis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter

def generate_data(num_points):
    random = np.random.randn(num_points)
    data = np.abs(random.cumsum())  # neneigiama suma
    return data

def create_date_range(start_date, end_date, num_points):
    dates = pd.date_range(start_date, end_date, periods=num_points)
    return dates

def create_line_plot(dates, data, end_date, line_colour1='red', line_colour2='blue', linewidth1=1, linewidth2=1, title=""):
    plt.figure(figsize=(12, 6))
    plt.title(title)
    plt.xlabel("Time")
    plt.ylabel("Price change")

    plt.plot(dates, data * 0.95, label='Daily', alpha=0.7, color=line_colour2, linewidth=linewidth2)
    plt.plot(dates, data, label='Monthly', color=line_colour1, linewidth=linewidth1)

    plt.gca().yaxis.set_major_formatter(StrMethodFormatter('${x:,.0f}')) # dolerio ženklas

    plt.legend(loc='upper left') # legenda - kairėje

    plt.grid(which='major', linestyle='-')  # keičiamas tinklelis (Major grid)
    plt.minorticks_on()
    plt.grid(which='minor', linestyle='--', alpha=0.2)  # keičiamas tinklelis (Minor grid)

    plt.xlim(dates[0], end_date)  # x ašies ribos

    plt.show()

num_points = 150
start_date = '2014-09-01'
end_date = '2024-01-29'

prices = generate_data(num_points)
dates = create_date_range(start_date, end_date, num_points)

create_line_plot(dates, prices, pd.to_datetime(end_date), line_colour1='red', line_colour2='blue', linewidth1=3, linewidth2=2, title='Randomly Generated Data from 2014-09-01 to 2024-01-29')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Nustatome pradines reikšmes
n_days = 365
n_months = 12
correction_factor = 0.9  # Korekcijos koeficientas
start_date = datetime.strptime('2022-09-01', '%Y-%m-%d')

# Generuojami atsitiktiniai dienos kainų skirtumai
daily = np.random.randn(n_days)

price_daily = abs(np.cumsum(daily)) # cumsum sumavimas

# Dienos duomenys konvertuojami į mėnesio
price_monthly = np.zeros(n_months)
for i in range(n_months):
    start_index = i * (n_days // n_months)
    end_index = (i + 1) * (n_days // n_months)
    price_monthly[i] = price_daily[end_index - 1]

x_days = np.arange(n_days) # x ašies reikšmės
x_months = np.arange(n_months) * (n_days // n_months)

price_monthly_corrected = price_monthly + (price_daily[-1] - price_monthly[-1]) * correction_factor # koreguojamos mėnesio kainos

plt.figure(figsize=(10, 5)) # grafikas

plt.plot(x_days, price_daily, label='Daily') # dieniniai duomenys

plt.plot(x_months, price_monthly_corrected, linestyle='-', color='red', label='Monthly') # mėnesio duomenys

plt.xticks(ticks=x_months, labels=[(start_date + timedelta(days=int(i))).strftime('%Y-%m') for i in x_months]) # nustatomi x ašies žymekliai

plt.title('Randomly Generated Data')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as dates
from matplotlib.dates import HourLocator, DateFormatter

# Sukuriama plot_price_data funkcija
def plot_price_data(ax, data, label, color, ylabel='Price', xlabel='Time'):
    ax.plot(data.index, data['Close'], label=label, color=color)
    ax.set_ylabel(ylabel, fontweight='bold')
    ax.set_xlabel(xlabel, fontweight='bold')
    ax.legend(loc='upper right')
    ax.grid(True)

# Pakeičiamas stilius į "bmh"
plt.style.use("bmh")

minute_data_ES_F['Hour'] = minute_data_ES_F.index.hour # naujas stulpelis 'Hour'

# Pasirenkami duomenys nuo 9:00 iki 16:00
session_data = minute_data_ES_F[(minute_data_ES_F['Hour'] >= 9) & (minute_data_ES_F['Hour'] < 16)].copy()

session_data.drop(columns=['Hour'], inplace=True) # pašalinamas stulpelis 'Hour'

# Duomenų imties keitimas kas minutę ir trūkstamų reikšmių užpildymas NaN
session_data = session_data.resample('1T').asfreq()

# Bendri grafikai (subplot)
fig = plt.figure(figsize=(14, 8))
gs = fig.add_gridspec(2, hspace=0)

ax1 = fig.add_subplot(gs[0])
ax2 = fig.add_subplot(gs[1], sharex=ax1)

fig.suptitle('E-Mini S&P 500 Mar 24 (ES=F). Trading Sessions (2024-02-13 – 2024-02-16)', fontweight='bold')

# ES=F close kainos (full periodo) vizualizavimas, naudojant sukurtą funkciją
plot_price_data(ax1, minute_data_ES_F, 'ES=F Close Price (Full Period)', 'green')

# Uždedamos vertikalios linijos viršutiniame grafike (00:00) ir pažymima data
for date, data in minute_data_ES_F.groupby(minute_data_ES_F.index.date):
    if date >= pd.Timestamp('2024-02-14').date():
        ax1.axvline(pd.Timestamp(date), color='gray', linestyle='--', linewidth=1)
        ax1.text(pd.Timestamp(date), ax1.get_ylim()[1], date.strftime('%Y-%m-%d'), rotation=0, ha='left', va='top', fontsize=9, color='gray')

# Pašalinamos žymos iš viršutinio grafiko x-ašies
ax1.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)

# ES=F close kainos vizualizavimas (nuo 9:00 iki 16:00), naudojant sukurtą funkciją
plot_price_data(ax2, session_data, 'ES=F Close Price (9:00 AM to 4:00 PM)', 'darkorange')

# x-ašyje pradedama numeruoti nuo 5:00 iki 19:00
start_time = pd.Timestamp(session_data.index[0].date()) + pd.Timedelta(hours=5)
end_time = pd.Timestamp(session_data.index[-1].date()) + pd.Timedelta(hours=19)
ax1.set_xlim(start_time, end_time)
ax2.set_xlim(start_time, end_time)

# Pakeičiamas datos formatas kuris rodomas x-ašyje
formatter = DateFormatter('%H:%M')
ax2.xaxis.set_major_formatter(formatter)

# x-ašies žymos pateikiamos naudojant 3-ų valandų intervalą (antrasis grafikas)
ax2.xaxis.set_major_locator(HourLocator(interval=3))

# Uždedamos vertikalios linijos apatiniame grafike (9:00 ir 16:00)
for date, data in session_data.groupby(session_data.index.date):
    ax2.axvline(pd.Timestamp(date) + pd.Timedelta(hours=9), color='gray', linestyle='--', linewidth=1)
    ax2.axvline(pd.Timestamp(date) + pd.Timedelta(hours=16), color='gray', linestyle='--', linewidth=1)

plt.xticks(rotation=25)
plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import pandas as pd

def data_10_years(ticker, start_date, end_date, file_name):
    data = yf.download(ticker, start=start_date, end=end_date)
    if isinstance(data.columns, pd.MultiIndex):
        data.columns = data.columns.get_level_values(0)

    cols = ['Open', 'High', 'Low', 'Close', 'Volume']
    data = data[cols]
    
    data.to_csv(file_name)
    data.columns.name = None
    return data

data_wheat = data_10_years("ZW=F", "2014-01-01", "2024-01-01", "yfinance_data_wheat.csv")
print("Wheat Futures Data:")
print(data_wheat.head().to_string())

data_corn = data_10_years("ZC=F", "2014-01-01", "2024-01-01", "yfinance_data_corn.csv")
print("\nCorn Futures Data:")
print(data_corn.head().to_string())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def calculate_avg_price(data):
    prices = data['Close'] # pasirenkama 'Close' kainos

    # Pagal mėnesius sugrupuojamos kainos ir suagreguojama dešimties metų laikotarpiu
    monthly_avg_prices = prices.groupby(prices.index.month).mean()

    overall_avg_price = monthly_avg_prices.mean() # apskaičiuojamas bendras vidurkis

    return monthly_avg_prices, overall_avg_price

def plot_seasonal_trend(monthly_avg_prices, overall_avg_price, name):
    plt.figure(figsize=(10, 6))

    # Laikotarpio, kai kaina nuolat augo, suradimas ir pavaizdavimas
    growing_period_start = None
    growing_period_end = None
    for i in range(len(monthly_avg_prices) - 1):
        if monthly_avg_prices.values[i + 1] > monthly_avg_prices.values[i]:
            if growing_period_start is None:
                growing_period_start = monthly_avg_prices.index[i]
        elif growing_period_start is not None:
            growing_period_end = monthly_avg_prices.index[i]
            plt.axvspan(growing_period_start, growing_period_end, color='goldenrod', alpha=0.5)
            growing_period_start = None

    plt.plot(monthly_avg_prices.index, monthly_avg_prices.values, linestyle='-', color='black', label='Monthly Average')
    plt.axhline(overall_avg_price, linestyle='--', color='black', label='Overall Average')

    plt.title('10 Year Seasonal Trend (2014-2024)')
    plt.suptitle(name + ' Futures, May-2', fontsize=14, fontweight='bold')
    plt.xlabel('Month', fontweight='bold')
    plt.ylabel('Price ($)', fontweight='bold')
    plt.xticks(range(1, 13), ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC'])
    plt.grid(True)
    plt.legend()
    plt.gca().set_facecolor('whitesmoke') # fono spalva
    plt.show()

# bendros vidutinės kviečių kainos ir grafikas
monthly_avg_prices_wheat, overall_avg_price_wheat = calculate_avg_price(data_wheat)
plot_seasonal_trend(monthly_avg_prices_wheat, overall_avg_price_wheat, 'Chicago SRW Wheat')

# bendros vidutinės kukurūzų kainos ir grafikas
monthly_avg_prices_corn, overall_avg_price_corn = calculate_avg_price(data_corn)
plot_seasonal_trend(monthly_avg_prices_corn, overall_avg_price_corn, 'Corn')
```
